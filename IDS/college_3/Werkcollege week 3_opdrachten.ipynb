## 1. 
**Importeer de dataset Mugello.csv** <br>
gebruik de optie index_col =0 <br>
**De dataset bevat rondetijden van coureurs in de formule 1 in 1 race in secondes**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df_raw = pd.read_csv('Mugello.csv', index_col=0)
df_raw.head()
```

## 2. 
**Zet de dataset om van een wide format naar een long format zodat je per ronde, per coureur de rondetijd krijgt.**

Long & wide formats: https://www.youtube.com/watch?v=gczclCnayhY&list=PLwkTCAI_gJjW10xaAJpdqeQx7EQgFolfG&index=13 

```python
df = pd.melt(df_raw, id_vars='LAP', var_name='Driver', value_name='Time')
df.head()
```

## 3. 
**Wie reed de snelste ronde?**<br> 
Data filtering: https://www.youtube.com/watch?v=gczclCnayhY&list=PLwkTCAI_gJjW10xaAJpdqeQx7EQgFolfG&index=13

Je kan sorteren:

```python
df.sort_values(by='Time', inplace=True, ascending=True)
df.head()
```

Of filteren op de laagste rondetijd: 

```python
fastest_time = df['Time'].min()
fastest_driver = df.loc[(df['Time'] == fastest_time)]

print(fastest_driver)
```

## 4. 
**Verwijder de NA waardes uit de dataset**



```python
df.dropna(inplace=True)
```

## 5. 
**Wat is de gemiddelde rondetijd?**

```python
avage_time = df['Time'].mean()
print(avage_time)
```

## 6. 
**Bereken de snelste rondetijd per coureur en sla deze op als variabele in het dataframe**<br>
**Bereken tevens het verschil van de rondetijd met de snelste ronde deze coureur en sla dit op in een nieuwe variabele, doe deze twee berekeningen in één assign functie**

Groupby instructies: https://www.youtube.com/watch?v=L_Ck90EGRpM&list=PLwkTCAI_gJjW10xaAJpdqeQx7EQgFolfG&index=16 <br>
Assign functies: https://www.youtube.com/watch?v=EVkxWknOo4w&list=PLwkTCAI_gJjW10xaAJpdqeQx7EQgFolfG&index=17

```python
# get fastest time for each driver put in new column
df['Fastest_time_per_driver'] = df.groupby('Driver')['Time'].transform('min')

# get difference between every lap and fastest lap per driver
df['Difference'] = df['Time'] - df['Fastest_time_per_driver']

df.head()
```

## 7. 
**Bereken de totale tijd gereden vanaf de eerste ronde per coureur. Maak een nieuwe kolom
om deze informatie op te slaan.**

```python
# add time every lap to get total time
df['Total_time'] = df.groupby('Driver')['Time'].transform('cumsum')
df.head()
```

## 8. 
**Zet de dataset op volgorde van totale tijd verstreken van klein naar groot**<br>
**Welke coureur lag er op kop in ronde 10?**

```python
# sort by sum from lowest to highest
df.sort_values(by='Total_time', inplace=True, ascending=True)

# get the fastest driver by sum in 10th lap
fastest_driver = df.loc[(df['LAP'] == 10)]

display(fastest_driver)
```
